# Интерактивная настройка PID-регулятора с помощью алгоритма Циглера-Никольса

## Задачи

### PID-регулятор

Измените моделирование, чтобы включить трехпараметрическое PID-управление формой

$$G_c(s) = K_c\left(1 + \frac{1}{\tau_Is} + \frac{\tau_Ds}{\frac{\tau_D}{N} s + 1}\right)$$

где $N = 10$.

## Функции передачи процесса

Динамика клапана:
$$G_v(s) = \frac{1}{2s + 1}$$

Динамика процесса:
$$G_p(s) = \frac{1}{5s + 1}$$

Динамика возмущений:
$$G_d(s) = \frac{1}{5s + 1}$$





```python
Gv = control.tf([1],[2,1])
Gp = control.tf([1],[5,1])
Gd = control.tf([1],[5,1])
print(Gv,Gp,Gd)
```

### Передаточная функция блока измерений

Измерение состоит из чистой временной задержки в 1 минуту, моделируемой передаточной функцией

$$G_m(s) = e^{-\tau_d s}$$

где $\tau_d = 1$. Библиотека control не поддерживает звенья запаздывания, моделируемые экспонентами, поэтому необходима аппроксимация. Аппроксимация Паде первого порядка задается формулой

$$e^{-\tau_d s} \approx \frac{-\frac{\tau_d}{2}s + 1}{\frac{\tau_d}{2}s + 1}$$

Это реализовано в библиотеке управления python с помощью функции `page(tdelay,norder)`, где `tdelay` - длина временной задержки, а `norder` - порядок приближения.

```python
tdelay = 1
norder = 1
num,den = control.pade(tdelay,norder)
Gm = control.tf(num,den)
print(Gm)
```

## Передаточные функции замкнутых систем

### Определения

Задача состоит в том, чтобы выбрать параметры для P, PI или PID-регулятора, которые приводят к хорошему отклику системы. Как мы определяем, является ли реакция системы хорошей или нет?

Ключом к этому анализу является рассмотрение набора из четырех откликов

| Отклик | Передаточная функция | Описание |
| :----------------:| :---------------: | :---------- |
| $r \rightarrow y$ | $H_{yr}$ | выходная реакция на изменение заданного значения (в идеале 1)  
| $d \rightarrow y$  | $H_{yd}$ | выходной отклик на помеху (в идеале 0) |
| $r \rightarrow u$ | $H_{ur}$ | управляемый отклик переменной на изменение заданного значения |
| $d \rightarrow u$ | $H_{ud}$ | управляемая переменная реакция на возмущение |

Только изучив все четыре, можно определить, хорошо ли регулятор справляется как с отслеживанием заданного значения, так и с подавлением помех, и что он обеспечивает хорошее управление выходом без чрезмерных или неприемлемых изменений в управляемой переменной.

Передаточный функции замкнутой системы обозначаются $H$, при этом индексы указывают на конкретное сопряжение системных входов и выходов. Индексы соответствуют индексам строк и столбцов в матричном уравнении

$$\left[\begin{array}{c} y \\ u \end{array}\right] =
\left[\begin{array}{cc} H_{yr} & H_{yd} \\ H_{ur} & H_{ud} \end{array}\right]
\left[\begin{array}{c} r \\ d \end{array}\right]$$

Для блок-схемы замкнутой системы, показанной выше, передаточные имеют вид

$$H_{yr} = \frac{G_p G_v G_c}{1 + G_p G_v G_c G_m}$$

$$H_{yd} = \frac{G_d}{1 + G_p G_v G_c G_m}$$

$$H_{ur} = \frac{G_c}{1 + G_c G_m G_p G_v}$$

$$H_{ud} = -\frac{G_c G_m G_d}{1 + G_c G_m G_p G_v}$$

### Интерактивная модель

$$G_c(s) = K_c\left(1 + \frac{1}{\tau_Is} + \frac{\tau_Ds}{\frac{\tau_D}{N} s + 1}\right)$$


```python
def sim(Kc, tauI, tauD):
    N = 10
    Gc =  Kc*(1 + control.tf([0,1],[tauI,0]) + control.tf([tauD,0],[tauD / N, 1]))
    
    Hyr = Gp*Gv*Gc/(1+Gp*Gv*Gc*Gm)
    Hyd = Gd/(1+Gp*Gv*Gc*Gm)
    Hur = Gc/(1+Gc*Gm*Gp*Gv)
    Hud = -Gc*Gm*Gd/(1+Gc*Gm*Gp*Gv)

    t = np.linspace(0,25,1000)

    plt.figure(figsize=(15,6))

    plt.subplot(2,2,1)
    y,t = control.step(Hyr,t)
    plt.plot(t,y)
    plt.plot(t,np.ones_like(y),'--',color ='b')
    plt.ylim(min(y) - 1,max(y) + 1)
    plt.title('Отклик на ступенчатый вход')
    plt.ylabel('y')

    plt.subplot(2,2,2)
    y,t = control.step(Hyd,t)
    plt.plot(t,y)
    plt.ylim(min(y) - 1,max(y) + 1)
    plt.plot(t,np.zeros_like(y),'--',color ='b')
    plt.title('Отклик на ступенчатое возмущение')
    plt.ylabel('y')

    plt.subplot(2,2,3)
    u,t = control.step(Hur,t)
    plt.plot(t,u)
    plt.ylim(min(u) - 1,max(u) + 1)
    plt.plot(t,np.zeros_like(u),'--',color ='b')
    plt.title('Отклик по управлению на ступенчатый вход')
    plt.ylabel('u')

    plt.subplot(2,2,4)
    u,t = control.step(Hud,t)
    plt.plot(t,u)
    plt.plot(t,np.ones_like(u) * (-1),'--',color ='b')
    plt.ylim(min(u) - 1,max(u) + 1)
    plt.title('Отклик по управлению на ступенчатое возмущение')
    plt.ylabel('u')

    plt.tight_layout()
```

```python
from ipywidgets import interact, fixed
interact(sim,Kc = (0.1,10,0.1),tauI=(0,25,0.1), tauD=(0,25,0.1))
```

## Процедуры настройки управления

Существуют правила настройки Пропорционального, Пропорционально-интегрального и Пропорционально-Интегрально-производного управления, которые помогают с выборам параметров. Процедуры для настройки Циглера-Николса и  Тиреуса-Люйбена следующие:

1. Пусть включена только пропорциональная компонента регулятора. Этого можно достичь, установив $\tau_D = 0$ и $\tau_I$ на максимально возможное значение.

2. Будем увеличивать коэффициент усиления $K_c$ до тех пор, пока отклик не примет вид гармонических колебаний. Такой отклик - это граница между устойчивой и неустойчивой реакцией системы. Значение $K_c$, при котором отклик приобретает форму гармонических колебаний обозначим $K_{cu}$, а период колебаний - $P_u$.

3. Получив значения для $K_{cu}$ и $P_u$, выберите настройки ПИД регулятора либо по Циглеру-Никольсу, либо по Тиреусу-Люйбену.

4. Проанализируйте отклик системы на характерные для её рабочих режимов сигналы. При необходимости перенастройте коэффициенты регулятора из общих соображений.

Чтобы продемонстрировать это, мы использовали интерактивную модель для определения значений параметров из п.2. Полученные значения: $K_{cu} = 8,1$ и $P_u = 8$.

```python
from ipywidgets import interact, fixed
interact(sim,Kc = fixed(8.1),tauI=fixed(1000), tauD=fixed(0))
```

### Циглер-Никольс (классический PID)

Один из наиболее известных наборов правил настройки был разработан Дж. Дж. Зиглером и Н.Б. Николсом в 1942 году, когда они работали в Taylor Instrument Company в Рочестере, штат Нью-Йорк. Правила кратко излагаются следующим образом

| Регулятор | $K_c$         | $\tau_I$          | $\tau_D$        |
| :-----: | :-----------: | :---------------: | :-------------: |
| P       | $0.5 K_{cu}$  | -                 | -               |
| PI      | $0.45 K_{cu}$ | $\frac{P_u}{1.2}$ | -               |
| PID     | $0.6 K_{cu}$  | $\frac{P_u}{2}$   | $\frac{P_u}{8}$ |

$$K_c = 0.6 K_{cu} = 0.6 \times 8.1 = 4.86 $$

$$\tau_I = \frac{P_u}{2} = \frac{8}{2} = 4$$

$$\tau_D = \frac{P_u}{8} = \frac{8}{8} = 1$$


```python
interact(sim,Kc = fixed(4.86), tauI=fixed(4),tauD=fixed(1));
```

### Тирей-Луйбен

Правила настройки Циглера-Никольса обычно считаются слишком агрессивными для большинства приложений управления технологическими процессами. В 1992 году, основываясь на совместных исследованиях химической компании Dow и Университета Лихай, Бьорн Шин и Уильям Луйбен предложили следующую модификацию традиционных правил настройки.

| Регулятор | $K_c$         | $\tau_I$     | $\tau_D$          |
| :-----: | :-----------: | :----------: | :---------------: |
| PI      | $0.31 K_{cu}$ | $2.2 P_u$    | -                 |
| PID     | $0.45 K_{cu}$  | $2.2 P_u$   | $\frac{P_u}{6.3}$ |

$$K_c = 0.45 K_{cu} = 0.45 \times 8.1 = 3.645 $$

$$\tau_I = 2.2P_u = 8 \times 2.2 = 17.6$$

$$\tau_D = \frac{P_u}{6.3} = \frac{8}{6.3} = 1.2698$$

```python
interact(sim,Kc = fixed(3.645), tauI=fixed(17.6),tauD=fixed(1.2698));
```

### Альтернативные правила настройки (из Wiki)

#### Интегральное правило Пессена

| Регулятор | $K_c$         | $\tau_I$     | $\tau_D$          |
| :-----: | :-----------: | :----------: | :---------------: |
| PID     | $0.7 K_{cu}$  | $0.4 P_u$   | $0.15 P_u$ |

$$K_c = 0.7 K_{cu} = 0.7 \times 8.1 =  5.67$$

$$\tau_I = 0.4P_u = 8 \times 0.4 = 3.2$$
$$\tau_D = 0.15P_u = 8 \times 0.15 = 1.2$$

```python
interact(sim,Kc = fixed(5.67), tauI=fixed(3.2),tauD=fixed(1.2));
```

#### Небольшой overshoot

| Регулятор | $K_c$         | $\tau_I$     | $\tau_D$          |
| :-----: | :-----------: | :----------: | :---------------: |
| PID     | $0.33 K_{cu}$  | $0.5 P_u$   | $0.33 P_u$ |


$$K_c = 0.33 K_{cu} = 0.33 \times 8.1 =  2.673$$

$$\tau_I = 0.5P_u = 8 \times 0.5 = 4$$

$$\tau_D = 0.33P_u = 8 \times 0.33 = 2.64$$

```python
interact(sim,Kc = fixed(2.673), tauI=fixed(4),tauD=fixed(2.64));
```

#### Без overshoot-а

| Регулятор | $K_c$         | $\tau_I$     | $\tau_D$          |
| :-----: | :-----------: | :----------: | :---------------: |
| PID     | $0.2 K_{cu}$  | $0.5 P_u$   | $0.33 P_u$ |

$$K_c = 0.2 K_{cu} = 0.2 \times 8.1 =  1.62$$

$$\tau_I = 0.5P_u = 8 \times 0.5 = 4$$

$$\tau_D = 0.33P_u = 8 \times 0.33 = 2.64$$

```python
interact(sim,Kc = fixed(1.62), tauI=fixed(4),tauD=fixed(2.64));
```

### Сравнение методов

С помощью графиков мы можем наблюдать, что Циглер-Никольс (классический PID), "Интеграл Пессена" быстрее приводят систему в стабильное состояние, но у них более высокий overshoot, чем у других методов. Правила настройки Тайреуса-Луйбена и "Небольшой overshoot" имеют минимальный overshoot, но для достижения стабильного состояния требуется больше времени. И последний метод "Без overshoot-a" соответствует своему названию, и он требует много времени.
В заключение можно сказать, что настройка Циглера–Николса приемлема для некоторых целей, но не оптимальна для всех задач. Достоинством метода является его простота, а основным недостатком — неучитывание требований к запасу устойчивости
