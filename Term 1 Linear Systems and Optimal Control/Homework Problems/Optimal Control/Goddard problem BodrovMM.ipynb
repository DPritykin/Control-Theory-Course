# Одномерная задача Годдарда
#### Задача поиска оптимального управления для максимизации подъема вертикально взлетающей ракеты при наличии влияния атмосферного давления и гравитационного поля земли в одномерном случае.


Движение ракеты задается системой обыкновенных дифференциальных уравнений:

$$
\left\{
\begin{array}
    \ \dot{h}=v \\
     m \dot{v} = -mg - Q + P \\
     \dot{m} = -u,   \ 0\leq u\leq u^{*}
\end{array}
\right.
$$

где $ P = \gamma u$ — реактивная тяга ракеты, $\gamma = const$ — относительная скорость истечения топлива.
$ Q = \frac{\rho v^{2}}{2}Sc = k v^{2}$ — сопротивление воздуха.

Далее будет введен фазовый вектор:

$$x =
\left(
    \begin{array}\\
       x_1\\
       x_2\\
       x_3 
    \end{array}   
\right)=
\left(
    \begin{array}\\
       h\\
       v\\
       m 
    \end{array}   
\right)
$$

С помощью него систему можно переписать следующим образом:

$$
\left\{
\begin{array}
    \ \dot{x_1}=x_2 \\
     \dot{x_2} = -g +\frac{\gamma u - Q(x_2)}{x_3} \\
     \dot{x_3} = -u,   
\end{array}
\right.
$$


Полученное начальное многообразие:
$$    \begin{array}\\
       h(0) = 0 \\
       v(0) = 0 \\
       m(0) = m_0
    \end{array}  
\qquad \rightarrow \qquad x=
\left(
    \begin{array}\\
       0\\
       0\\
       m_0 
    \end{array}   
\right)
\qquad \rightarrow \qquad
x(t_0) \in A = \left\{
    \begin{array}\\
       x_1 = h = 0 \\
       x_2 = v = 0 \\
       x_3-m_0 = 0\\
       t = 0
    \end{array}   
\right.
$$

$m_0$ - начальная масса ракеты с топливом.

Ограничения на управление:
   $\ 0\leq u\leq u^{*}$,
   где $u^*$- максимальный расход топлива.
   
Если принять собственную массу ракеты за $0<m_1<m_0$, тогда можно получить условие, формирующее терминальное многообразие:

$$\left\{
\begin{array}
 \ x_3(t_f)=m_1 
\end{array}
\right\} = B
$$

Терминальное многообразие:
$$    
\quad   
x(t_f) \in B = \left\{
    \begin{array}\\
       x_1 = h^*>0  \\
       x_2 = v^* - любая \\
       x_3 = m_1 \\
       t = t_f
    \end{array}   
\right.
$$

Требуется обеспечить максимальное значение высоты $h$ к моменту времени $t_f$, таким образом, функционал задачи:

$$\Phi(x(t_f)) = x_1(t_f) \rightarrow \max_{u}\\
I(u)=\Phi_0(x(t_f))=-x_1(t_f) \rightarrow \min_{u}
$$

Воспользуемся принципом максимума Понтрягина. Для этого необходимо выписать функцию Гамильтона-Понтрягина:

$$
H(x,p,u,t) = p_1x_2 + p_2(-g+\frac{\gamma u - Q}{x_3}) + p_3(-u) = H_0(p,x)+H_1(p,x)\cdot u, \\
H_1 =\frac{p_2\gamma}{x_3}-p_3, \\
H_0 = p_1x_2-p_2(\frac{Q}{x_3}+g)
$$

Дифференциальные уравнения на сопряженные переменные будут выглядеть следующим образом:

$$
\left\{
    \begin{array}\\
       \dot{p_1} = -\frac{\partial H}{\partial x_1} = 0 \\
       \dot{p_2} = -\frac{\partial H}{\partial x_2} = -p_1 + \frac{p_2}{x_3}\frac{\partial Q}{\partial x_2} \\
       \dot{p_3} = -\frac{\partial H}{\partial x_3} =  p_2\frac{\gamma u - Q}{x_3^2}
    \end{array}   
\right. 
$$


Условие трансверсальности на правом конце :
$$p(t_f) = \frac{\partial \Phi}{\partial 
\left(
    \begin{array}\\
       x \\
       v \\
       m
    \end{array}   
\right)}  = 
\left(
    \begin{array}\\
       1 \\
       0 \\
       0
    \end{array}   
\right).$$


Чтобы удовлетворить условиям принципа максимума, управление может принимать следующие значения : 

$$
\quad
u(t) = \left\{
               \begin{array}\\ 
               u^{*},\quadпри\quad H_1>0\quad \\
               0,\quadпри\quad H_1<0\quad \\
               любое\quadв\quadпределах\quad[0,u^{*}],\quadпри\quad H_1=0\end{array}
 \right.
 $$
 
На участке с неизвестным управлением $H_0(t)=H-H_1u\equiv0$ : $H_0(t)= p_1x_2-p_2(\frac{Q}{x_3}+g) = 0 \mapsto p_2=\frac{x_2}{\frac{Q}{x_3}+g}$

 Рассмотрим производную множителя $H_1$:
 
$$ \dot{H_1}= \frac{\dot p_2 \gamma}{x_3} - \frac{\dot x_3 \gamma p_2}{x_3^2}-\dot p_3 = (-1+\frac{p_2}{x_3}\frac{\partial Q}{\partial x_2})\frac{\gamma}{x_3}+\frac{up_2\gamma}{x_3^2}+p_2\frac{Q-\gamma u}{x_3^2} = \frac{\gamma}{x_3}(\frac{\partial Q}{\partial x_2}\frac{p_2}{x_3}-1)+\frac{Qp_2}{x_3^2} = \frac{1}{x_3^2}[(Q+\gamma\frac{\partial Q}{\partial x_2})p_2-\gamma x_3]\equiv0
$$

Ввиду того, что $\frac{1}{x_3^2}>0$ при любом $t$, получается, что $[(Q+\gamma\frac{\partial Q}{\partial x_2})p_2-\gamma x_3]=0$. 

Подстановкой выражения для $p_2$ получим: $[(Q+\gamma\frac{\partial Q}{\partial x_2})\frac{x_2}{\frac{Q}{x_3}+g}-\gamma x_3]=0$


$$ \begin{array}\\
-\gamma kx_2^2+\gamma gx_3-kx_2^3=0\\
\quad \\
m = k\frac{v^3+\gamma v^2}{\gamma g}\end{array}
$$

Продифференцируем второй раз:

$$\ddot H_1 = -\frac{2}{x_3^3}[(Q+\gamma\frac{\partial Q}{\partial x_2})p_2-\gamma x_3]+\frac{1}{x_3^2}[\dot p_2(Q+\gamma\frac{\partial Q}{\partial x_2})-\gamma\dot x_3+p_2(\frac{\partial Q}{\partial x_2}+\gamma\frac{\partial^2 Q}{\partial x_2^2})\dot x_2]= 0 + \frac{1}{x_3^2}[\dot p_2(Q+\gamma\frac{\partial Q}{\partial x_2})-\gamma\dot x_3+p_2(\frac{\partial Q}{\partial x_2}+\gamma\frac{\partial^2 Q}{\partial x_2^2})\dot x_2]=\frac{\gamma}{x_3^2}[1+\frac{p_2}{x_3}(\frac{\partial Q}{\partial x_2}+\gamma\frac{\partial^2 Q}{\partial x_2^2})]u + \frac{1}{x_3^2}[(Q+\gamma\frac{\partial Q}{\partial x_2})(\frac{\partial Q}{\partial x_2}\frac{p_2}{x_3}-1)-p_2(\frac{Q+ gx_3}{x_3})(\frac{\partial Q}{\partial x_2}+\gamma\frac{\partial^2 Q}{\partial x_2^2})]\equiv0
$$

Подстановкой $Q=kx_2^2$ и $p_2=\frac{x_2}{\frac{Q}{x_3}+g}$ получим:

$$\gamma u \frac{x_2^2+4\gamma x_2 +2\gamma^2}{kx_2(x_2+2\gamma)}-\frac{kx_2^4+4k\gamma x_2^3 +2k\gamma^2x_2^2+2g\gamma x_2x_3+2g\gamma^2x_3}{kx_2(x_2+2\gamma)}=0
$$

Выразим $u$:

$$u^{s}=\frac{k}{\gamma}\frac{kx_2^4+4k\gamma x_2^3 +2k\gamma^2x_2^2+2g\gamma x_2x_3+2g\gamma^2x_3}{x_2^2+4\gamma x_2 +2\gamma^2}$$

Получено уравнение поверхности оптимальных управлений при  $H_1=0$ (особой поверхности). 
В зависимости от начальных условий оптимальная траектория либо достигнет особой
поверхности, либо нет.
Если движение происходит по особому участку, то $u=u^s$.

Выше особой поверхности $H_1 > 0$, а ниже — $H_1 < 0$.

Получается, что ракета движется с максимальной тягой до касания особой поверхности, после чего ее управление начинает удовлетворять уравнению поверхности.


$$
\quad
u(t) = \left\{
               \begin{array}\\ 
               u^{*},\quadпри\quad H_1>0\quad \\
               0,\quadпри\quad H_1<0\quad \\
               u^{s},\quadпри\quad H_1=0\end{array}
 \right.
 $$
 
 Участок, где $u=0$ не интересен для задачи, т.к. при нем наблюдается падение ракеты. Оптимальное управление приобретает вид:
 
$$
\quad
u(t) = \left\{
               \begin{array}\\ 
               u^{*},\quadпри\quad t\in [0,t^*]\quad \\
               u^{s},\quadпри\quad t\in (t^*,t_f]\end{array}
 \right.
 $$
 
 Также возможен вырожденный случай, при котором поверхность не достигается, соответственно $t^*$ находится позже $t_f$.
 
Первая часть движения описывается следующей системой дифференциальных уравнений:
 
$$
\left\{
\begin{array}
    \ \dot{x_1}=x_2 \\
     \dot{x_2} = -g +\frac{\gamma u^* - Q(x_2)}{x_3} \\
     \dot{x_3} = -u^*,  u^* = const 
\end{array}
\right.
$$

Вторая (если присутствует):

$$
\left\{
\begin{array}
    \ \dot{x_1}=x_2 \\
     \dot{x_2} = -g +\frac{\gamma u^s(x_2,x_3) - Q(x_2)}{x_3} \\
     \dot{x_3} = -u^s(x_2,x_3)
\end{array}
\right.
$$

```python
import numpy as np # Импорт необходимых пакетов
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import scipy.integrate as sci
```

```python
class Parameters(object): # Класс параметров
    pass

# Физические параметры
param = Parameters()  # объект класса параметров
param.m1 = 3      # масса ракеты без топлива
param.m0 = 10    # начальная масса
param.k = 0.7     # обобщенная константа атмосферного сопротивления
param.gamma = 1000  # скорость исходящего потока топлива
param.u_max = 0.1 
param.g = 9.81

# Параметры симуляции
sim_param = Parameters()  # объект класса параметров
sim_param.t0 = 0          # время начала симуляции
sim_param.t1 = 70       # время окончания симуляции
sim_param.dt = 0.01       # шаг по времени
```

```python
tt = np.arange(sim_param.t0, sim_param.t1, sim_param.dt) # Задаем временную шкалу

# Решение для участка взлета с максимальным управлением
def func(t,y): # Задаем функцию, возвращающую значения правых частей системы ДУ
    return [y[1],                                                           # dx_1/dt
            -param.g+ ((param.gamma*param.u_max)-param.k*(y[1]**2))/(y[2]), # dx_2/dt
            - param.u_max]                                                  # dx_3/dt

x0 = [0,0,param.m0]  # Столбец начальных условий ( h=0, v=0, m=m0)

sollution = sci.solve_ivp(func,(sim_param.t0,sim_param.t1),x0) # Решаем систему ДУ на временном отрезке [t0,t1]

#sollution
```

```python
# Графики для наглядности решения

fig, (ax0, ax1, ax2) = plt.subplots(nrows=1, ncols=3, sharex=True, figsize=(12, 4))

ax0.set_title('h(t) для первой системы')
ax0.errorbar(sollution.t,sollution.y[0].T)
ax0.grid()
ax0.set_xlabel('время (с)')
ax0.set_ylabel('высота подъема (м)')

ax1.set_title('v(t) для первой системы')
ax1.plot(sollution.t,sollution.y[1].T)
ax1.grid()
ax1.set_xlabel('время (с)')
ax1.set_ylabel('скорость (м/с)')

ax2.set_title('m(t) для первой системы')
ax2.plot(sollution.t,sollution.y[2].T)
ax2.grid()
ax2.set_xlabel('время (с)')
ax2.set_ylabel('масса ракеты (кг)')

fig.suptitle('Решение первой системы на [0,70]')
plt.show()
```

```python
Mass = sollution.y[2].T # Выделяем вектор масс - решений первой системы
Vel = sollution.y[1].T # Выделяем вектор скоростей - решений первой системы

def f(v): # Формула массы, выраженной через скорость для особой поверхности
    m = (param.k*v**3+param.gamma*param.k*v**2)/(param.gamma*param.g)
    return m

# Далее ищем точку, в которой решение первой системы пересекает особую поверхность, чтобы перевести движение ракеты на нее.

ind1 = 0 # Индекс, фиксирующий точку пересечения

for i in range(len(Mass)-1): # Цикл, ищущий точку пересечения в векторе решений
    if f(Vel[i])< Mass[i] and f(Vel[i+1])>= Mass[i+1]:
        print('Время пересечения особой поверхности:', sollution.t[i].round(3), 'с\n', # Вывод контрольной точки
              'Скорость ракеты:',Vel[i].round(3), 'м/с\n',
              'Масса ракеты:', Mass[i].round(3), 'кг\n')
        ind1 = i # Фиксация контрольной точки
        break

M = np.concatenate((Mass[:i+2],f(Vel[:i+1])[::-1])) # Формируем объединенную траекторию: 
V = np.concatenate((Vel[:i+2],Vel[:i+1][::-1]))     # до поверхности движение по решению системы ДУ
                                                    # далее по поверхности
# Также в массив М сразу заносятся рассчитанные значения массы, на основе уравнения поверхности.

# Построение графиков

fig, ax = plt.subplots(figsize=(8, 4))

line1, = ax.plot(Vel,Mass, label='Движение при максимальном управлении')
line1.set_dashes([2, 2]) 
line2, = ax.plot(Vel,f(Vel), dashes=[6, 2], label='Движение по особой поверхности')
line3, = ax.plot(V,M, label ='Решение')

fig.suptitle('VM диаграмма полета ракеты')
ax.set_xlabel('скорость (м/с)')
ax.set_ylabel('масса ракеты (кг)')
ax.grid()
ax.legend()
plt.show()
```

```python
# Также необходимо отсечь от решения участок, где масса приравнивается к массе ракеты без топлива(он нас не интересует)
ind2 = 0 # Индекс, фиксирующий совпадение масс
for i in range(len(M)-1): # Цикл, ищущий точку пересечения в векторе решений
    if M[i]> param.m1 and M[i+1]<= param.m1:
        print('Время истощения запасов топлива:', sollution.t[i].round(3), 'с\n', # Вывод контрольной точки
              'Скорость ракеты:',Vel[i].round(3), 'м/с\n',
              'Масса ракеты:', Mass[i].round(3), 'кг\n')
        ind2 = i # Фиксация контрольной точки
        break

# Построение обновленной VM диаграммы

fig, ax = plt.subplots(figsize=(8, 4))

line3, = ax.plot(V[:ind2],M[:ind2], label ='Решение')

fig.suptitle('VM диаграмма полета ракеты')
ax.set_xlabel('скорость (м/с)')
ax.set_ylabel('масса ракеты (кг)')
ax.grid()
ax.legend()
plt.show()
```

```python
T = sollution.t # Заносим временную шкалу в более удобный массив
X_max = sollution.y[0].T # Получаем функцию высоты - решение первой системы
X = X_max[:ind1] # Создаем массив для решения с включением движения по поверхности

C = X[ind1-1] - V[ind1-1]*T[ind1-1] # Считаем константу интегрирования

for i in range(ind1,ind2,1): # Дополняем массив решениями с особой поверхности
    X=np.append(X,V[i]*T[i]+C)
# Построение графика

fig, ax = plt.subplots(figsize=(8, 4))

line1, = ax.plot(T[:ind2],X[:ind2], label ='Решение')

ax.vlines(T[list(X).index(max(X[:ind2]))], X.min(), X.max(),color='black',linestyle = ':')


fig.suptitle('График высоты полета ракеты до истощения запаса топлива')
ax.set_xlabel('время (с)')
ax.set_ylabel('высота полета (м)')
ax.grid()
ax.legend()
plt.show()

print("Максимальная высота полета:",max(X[:ind2]),"м") # Выводим значение вершины траектории
```

```python
#Составление графика оптимального управления

def us(m,v,params): # Вводим функцию подсчета управления на особой поверхности
    us = params.k/params.gamma*(params.k*v**4+4*params.k*params.gamma*v**3+2*params.k*(params.gamma**2)*v**2+2*params.g*params.gamma*v*m+2*params.g*(params.gamma**2)*m)/(v**2+4*params.gamma*v+2*params.gamma**2)
    return us

fig, ax = plt.subplots(figsize=(12, 4))

line1, = ax.plot(T[:ind1],[param.u_max]*len(T[:ind1]), label ='u*')
line2, = ax.plot(T[ind1-1:ind2],us(M[ind1-1:ind2],V[ind1-1:ind2],param), label ='Управление на особой поверхности')

ax.vlines(T[ind1-1], 0, param.u_max,color='black',linestyle = ':')
ax.vlines(T[ind2-1], 0, us(M[ind2-1],V[ind2-1],param),color='black',linestyle = ':')

fig.suptitle('Оптимальное управление')
ax.set_xlabel('время (с)')
ax.set_ylabel('управление')
ax.grid()
ax.legend()
plt.show()
```
