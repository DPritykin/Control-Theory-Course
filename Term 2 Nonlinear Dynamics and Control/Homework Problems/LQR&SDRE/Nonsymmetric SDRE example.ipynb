# State-Dependent Riccati Equation

Рассматривается система управления, модель которой представляется в виде
$$
\dot{\mathbf{x}}=\mathbf{f}(\mathbf{x})+\mathbf{B}(\mathbf{x}) \cdot \mathbf{u}, \quad \mathbf{x} \in \mathbb{R}^n, \quad u \in \mathbb{R}^m
$$
$\mathbf{x}=0$ - равновесие, то есть $\mathbf{f}(0)=0$.
Цель: найти управление $\mathbf{u}(t)$, минимизирующее функционал
$$
J=\frac{1}{2} \int_0^{\infty} \mathbf{x}^T(t) \mathbf{Q}(x) \mathbf{x}(t)+\mathbf{u}(t)^T \mathbf{R}(x) \mathbf{u}(t) d t
$$
где $\mathbf{Q} \geq 0, \mathbf{R}>0$
Закон управления ищется в виде
$$
\mathbf{u}(t)=\mathbf{k}(\mathbf{x})=-\mathbf{K}(\mathbf{x}) \mathbf{x}, \quad \mathbf{k}(0)=0
$$
Можно показать, что при условии $\mathbf{f}(0)=0$ и $(f) \in C^1\left(\mathbb{R}^n\right)$ всегда найдётся такая $\mathbf{A}(\mathbf{x})$, что
$$
\mathbf{f}(\mathbf{x})=\mathbf{A}(\mathbf{x}) \cdot \mathbf{x}
$$
Таким образом, исходая система может быть записана в виде:
$$
\dot{\mathbf{x}}=\mathbf{A}(\mathbf{x}) \cdot \mathbf{x}+\mathbf{B}(\mathbf{x}) \cdot \mathbf{u}
$$
эти уравнения имеют линейную структуру, но матрицы $\mathbf{A}$ и $\mathbf{B}$ завистя от вектора состояния $\mathbf{x}$. В литературе такое представление называют SDC (statedependent components).
Применение к такой "обобщённо-линейной" системе методов управления линейными системами приводит к нахождению законов управления вида $\mathbf{u}(t)=-\mathbf{K}(\mathbf{x}) \mathbf{x}$, что даёт матрицу системы с отрицательной обратной связью вида
$$
\mathbf{A}_{C L}(\mathbf{x})=\mathbf{A}(\mathbf{x})-\mathbf{B}(\mathbf{x}) \mathbf{K}(\mathbf{x})
$$
которая должна быть гурвицевой в каждой точке.
SDRE
применяя к системе алгоритм для задачи о линейно-квадратичном регуляторе, получим:
$$
\mathbf{u}(t)=\mathbf{R}^{-1}(\mathbf{x}) \mathbf{B}^T(\mathbf{x}) \mathbf{P}(\mathbf{x}) \cdot \mathbf{x}
$$
где $\mathbf{P}(\mathbf{x})$ - определённое единственным образом, симметрическое решение уравнения Риккати с коэффициентами, зависящими от вектора состояния:
$$
\mathbf{P}(\mathbf{x}) \mathbf{A}(\mathbf{x})+\mathbf{A}^T(\mathbf{x}) \mathbf{P}(\mathbf{x})-\mathbf{P}(\mathbf{x}) \mathbf{B}(\mathbf{x}) \mathbf{R}^{-1}(\mathbf{x}) \mathbf{B}^T(\mathbf{x}) \mathbf{P}(\mathbf{x})+\mathbf{Q}(\mathbf{x})=0
$$

# Решение задачи


Рассмотрим задачу управления нелинейной системой второго порядка:
<!-- \begin{equation}\label{sys}
    \begin{cases}
        \dot x_1 =  x_2 \\
        \dot x_2 =  -x_1 e^{x_1}+\frac{1}{2} x_2^2+e^{x_1} u
    \end{cases},
    \tag{1}
\end{equation} -->
$$
\left\{\begin{array}{l}

\end{array}\right.
$$

$\mathbf{Q}(\mathbf{x}) = diag \{0,1\} $, $\mathbf{R}(\mathbf{x}) = 1$.

$\textbf{Цель задачи:}$ найти управление $\mathbf{u}(t)$, минимизирующее функционал $J(\mathbf{x_0}, \mathbf{u})$

Матрица коэффициентов, зависящая от состояния (SDC matrices):
\begin{equation}
    \mathbf{A}= 
    \begin{bmatrix}
    0 & 1\\
    -e^{x_1} & \frac{1}{2}x_2
    \end{bmatrix}
\end{equation}


Решая задачу оптимального управления, выбираем матрицу $\mathbf{B}(\mathbf{x})$ как

\begin{equation}
    \mathbf{B} = 
    \begin{bmatrix}
    0\\
    e^{x_1}
    \end{bmatrix}  
\end{equation}


### Согласно статье, оптимальный закон управления следующий:
$$
u^* = -x_2
$$
С соотвествующей функцией $V(x)$:
$$
V(\mathbf{x})=\frac{1}{2}\left(x_1^2+x_2^2 e^{-x_1}\right)
$$

# В статье сразу приведено 3 матрицы $\mathbf{P}(\mathbf{x})$
Одна из которых симметричная поточечно неотрицательно определенная:
$$
\mathbf{P}_1(\mathbf{x})=\left[\begin{array}{cc}
e^{x_1} & 0 \\
0 & 1
\end{array}\right] p(\mathbf{x})
$$

и 2 несимметричных:
$$
\begin{aligned}
&\mathbf{P}_2(\mathbf{x})=\left[\begin{array}{cc}
c_1 & -\frac{1}{2}\left(c_1 x_2 e^{-x_1}+1-c_1^2\right) \\
-2 e^{-x_1} & x_2 e^{-2 x_1}-c_1 e^{-x_1}
\end{array}\right] \\
&\mathbf{P}_3(\mathbf{x})=\left[\begin{array}{cc}
c_1 & -\frac{1}{2}\left(c_1 x_2 e^{-x_1}+1-c_1^2\right) \\
0 & c_1 e^{-x_1}
\end{array}\right]
\end{aligned}
$$

, где  $p(\mathbf{x}) \triangleq \frac{1}{2}\left[x_2+\left(x_2^2+4 e^{2 x_1}\right)^{1 / 2}\right] e^{-2 x_1}$, и $c_1$ произвольная константа.

### Однако только $\mathbf{P}_3(\mathbf{x})$ удовлетворяет условию симметричности с $c_1 = 1$, которое было специально сформулировано в статье для случая нахождения несимметричного решения SDRE:
$$
P_{i j}(\mathbf{x})+\sum_{k=1}^n \frac{\partial P_{i k}(\mathbf{x})}{\partial x_j} x_k=P_{j i}(\mathbf{x})+\sum_{k=1}^n \frac{\partial P_{j k}(\mathbf{x})}{\partial x_i} x_k
$$
Она, соответственно, является решением уравнения в несимметричном случае: 
$$
\begin{aligned}
&\mathbf{P}^T(\mathbf{x}) \mathbf{A}(\mathbf{x})+\mathbf{A}^T(\mathbf{x}) \mathbf{P}(\mathbf{x})-\mathbf{P}^T(\mathbf{x}) \mathbf{B}(\mathbf{x}) \mathbf{R}^{-1}(\mathbf{x}) \mathbf{B}^T(\mathbf{x}) \mathbf{P}(\mathbf{x}) +\mathbf{Q}(\mathbf{x})=\mathbf{0}
\end{aligned}
$$

### Далее будет показано, что при удовлетворении данного условия, оптимальный закон управления сходится с законом, полученным в случае симметричного решения SDRE

```python
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp 
```

```python
# начальные условия x
x0 = np.array([0.5, 0.5])

```

```python
# решение алгебраического уравнения Риккати
def lqr_fun(A, B, Q, R):
    P = np.matrix(sci.linalg.solve_continuous_are(A, B, Q, R)) 
    K = -np.matrix(sci.linalg.inv(R) * (B.T * P))
 
    eigVals, eigVecs = sci.linalg.eig(A + B * K)
 
    return K, P, eigVals
# обычный sdre
def u_sdre(x):
    A = np.array([[0., 1.],  
                  [-np.exp(x[0]), (1/2) * x[1]]])

    B = np.array([[0.], [np.exp(x[0])]])
    
    Q = np.array([[0.,0.],
                  [0., 1.]])
    
    R = np.eye(1)
    K, _, _ = lqr_fun(A, B, Q, R)
    u = np.dot(K, x).item()

    return u
# sdre для случая P_3
def u_sdre_nonsymmetric(x):
    A = np.array([[0., 1.],  
                  [-np.exp(x[0]), (1/2) * x[1]]])

    B = np.array([[0.], [np.exp(x[0])]])
    
    Q = np.array([[0.,0.],
                  [0., 1.]])
    P = np.matrix([[1., -(1/2) * (x[1] * np.exp(-x[0]))],  
                  [0, np.exp(-x[0])]])
    R = np.eye(1)
    K = -np.matrix(sci.linalg.inv(R) * (B.T * P))
    
    u = np.dot(K, x).item()
    
    return u
# sdre для случая P_2
def u_sdre_nonsymmetric_bad(x,c):
    A = np.array([[0., 1.],  
                  [-np.exp(x[0]), (1/2) * x[1]]])

    B = np.array([[0.], [np.exp(x[0])]])
    
    Q = np.array([[0.,0.],
                  [0., 1.]])

    P = np.matrix([[c, -(1/2)*(c*x[1]*np.exp(-x[0]) + 1 - c*c)],  
                  [-2*np.exp(-x[0]), x[1]*np.exp(-2*x[0]) - c*np.exp(-x[0])]])
    R = np.eye(1)
    K = -np.matrix(sci.linalg.inv(R) * (B.T * P))
    u = np.dot(K, x).item()
    
    return u
# sdre для случая P_1
def u_sdre_symmetric_bad(x):
    A = np.array([[0., 1.],  
                  [-np.exp(x[0]), (1/2) * x[1]]])

    B = np.array([[0.], [np.exp(x[0])]])
    
    Q = np.array([[0.,0.],
                  [0., 1.]])
    a = x[1] + np.sqrt(x[1]*x[1] + 4*np.exp(2*x[0]))
    P = np.matrix([[(1/2) * np.exp(-x[0]) * a, 0.],  
                  [0., (1/2)*np.exp(-2*x[0]) * a]])
    R = np.eye(1)
    K = -np.matrix(sci.linalg.inv(R) * (B.T * P))
    u = np.dot(K, x).item()
    
    return u

def rhs_symmetric_bad(t, x):
  return np.array([ x[1], -x[0]*np.exp(x[0]) + (1/2)*x[1]*x[1] + np.exp(x[0])*u_sdre_symmetric_bad(x) ])

def rhs_nonsymmetric_bad(t, x):
  return np.array([ x[1], -x[0]*np.exp(x[0]) + (1/2)*x[1]*x[1] + np.exp(x[0])*u_sdre_nonsymmetric_bad(x,10) ])

def rhs(t, x):
  return np.array([ x[1], -x[0]*np.exp(x[0]) + (1/2)*x[1]*x[1] + np.exp(x[0])*u_sdre(x) ])

def rhs_nonsymmetric(t, x):
  return np.array([ x[1], -x[0]*np.exp(x[0]) + (1/2)*x[1]*x[1] + np.exp(x[0])*u_sdre_nonsymmetric(x) ])

t0 = 0.
tf = 10.
t_eval = np.arange(t0, tf, 0.01)
```

#Для начала самостоятельно найдем оптимальный закон управления и сравним его с законом, получающимся в случае несимметричного решения SDRE, удовлетворяющего условию симметричности, приведенному выше

```python

sol = solve_ivp(rhs, (t0, tf), x0, t_eval = t_eval)
x = sol.y.T
t = sol.t

sol_nonsymmetric = solve_ivp(rhs_nonsymmetric, (t0, tf), x0, t_eval = t_eval)
x_nonsymmetric = sol_nonsymmetric.y.T
t_nonsymmetric = sol_nonsymmetric.t
```

# Иллюстрируем решение

```python
fig = plt.figure(figsize=(35,15))


ax1 = fig.add_subplot(2,3,1)
ax1.plot(t, x[:,0],t_nonsymmetric, x_nonsymmetric[:,0])
ax1.set_xlabel('Время, с')
ax1.set_ylabel('x1')
ax1.legend(['symmetric', "nonsymmetric"])
ax1.set_title('Зависимость x1 от времени')
ax1.grid()

ax2 = fig.add_subplot(2,3,2)
ax2.plot(t,x[:,1],t_nonsymmetric, x_nonsymmetric[:,1])
ax2.set_xlabel('Время, с')
ax2.set_ylabel('x2')
ax2.legend(['symmetric', "nonsymmetric"])
ax2.set_title('Зависимость x2 от времени')
ax2.grid()


ax3 = fig.add_subplot(2,3,3)
ax3.plot(x[:,0], x[:,1],x_nonsymmetric[:,0], x_nonsymmetric[:,1])
ax3.set_xlabel('x1')
ax3.set_ylabel('x2')
ax3.legend(["Symmetric SDRE", "Nonsymmetric SDRE"])
ax3.set_title('Фазовый портрет')
ax3.grid()
```

# Видно, как накладываются друг на друга кривые

# Теперь сравним предыдущие 2 случая со случаем, когда $P(x)$ - симметричная поточечно неотрицательно определенная

```python
sol_symmetric_bad = solve_ivp(rhs_symmetric_bad, (t0, tf), x0, t_eval = t_eval)
x_symmetric_bad = sol_symmetric_bad.y.T
t_symmetric_bad = sol_symmetric_bad.t
```

```python
fig = plt.figure(figsize=(35,15))


ax1 = fig.add_subplot(2,3,1)
ax1.plot(t, x[:,0],t_nonsymmetric, x_nonsymmetric[:,0],t_symmetric_bad, x_symmetric_bad[:,0])
ax1.set_xlabel('Время, с')
ax1.set_ylabel('x1')
ax1.legend(['symmetric', "nonsymmetric", "symmetric_bad"])
ax1.set_title('Зависимость x1 от времени')
ax1.grid()

ax2 = fig.add_subplot(2,3,2)
ax2.plot(t,x[:,1],t_nonsymmetric, x_nonsymmetric[:,1],t_symmetric_bad, x_symmetric_bad[:,1])
ax2.set_xlabel('Время, с')
ax2.set_ylabel('x2')
ax2.legend(['symmetric', "nonsymmetric", "symmetric_bad"])
ax2.set_title('Зависимость x2 от времени')
ax2.grid()


ax3 = fig.add_subplot(2,3,3)
ax3.plot(x[:,0], x[:,1],x_nonsymmetric[:,0], x_nonsymmetric[:,1], x_symmetric_bad[:,0], x_symmetric_bad[:,1])
ax3.set_xlabel('x1')
ax3.set_ylabel('x2')
ax3.legend(["Symmetric SDRE", "Nonsymmetric SDRE", "Symmetric_bad SDRE"])
ax3.set_title('Фазовый портрет')
ax3.grid()
```

# Можно заметить, что в данном случае кривая полностью накладывается на кривую, которую мы получили вручную (без решений, предоставленных в статье)

```python
sol_nonsymmetric_bad = solve_ivp(rhs_nonsymmetric_bad, (t0, tf), x0, t_eval = t_eval)
x_nonsymmetric_bad = sol_nonsymmetric_bad.y.T
t_nonsymmetric_bad = sol_nonsymmetric_bad.t
```
